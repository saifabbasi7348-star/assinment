```python
# Take input from the user
n = int(input("Enter a number: "))

print("The first", n, "terms of the series are:")

# Generate the series
for i in range(1, n + 1):
    print(i ** 2, end=" ")
```

```python
# Take input from the user
num = int(input("Enter a number: "))

print("Multiplication Table of", num)

# Print the table from 1 to 10
for i in range(1, 11):
    print(num, "x", i, "=", num * i)
```

```python
# Take input from the user
n = int(input("Enter a number: "))

print("Odd numbers between 1 and", n, "are:")

# Print odd numbers
for i in range(1, n + 1, 2):
    print(i, end=" ")
```

```python
# Initialize the variable
num = 20

# Print numbers from 20 to 1
while num >= 1:
    print(num)
    num = num - 1
```

```python
# Take input from the user
n = int(input("Enter a number: "))

# Initialize the counter
i = 1

# Print numbers from 1 to n
while i <= n:
    print(i)
    i = i + 1
```

```python
# Initialize variables
count = 1
total = 0

# Repeat 5 times
while count <= 5:
    num = int(input("Enter a number: "))
    total = total + num
    count = count + 1

# Display the final total
print("The final running total is:", total)
```

```python
# Create the list
clean_countries = ["Bangladesh", "India", "Pakistan", "China"]

# Repeat until a valid country is entered
while True:
    country = input("Enter a country name: ")

    if country in clean_countries:
        print(country, "is in clean countries list")
        break
    else:
        print("Country not found. Try again.")
```
